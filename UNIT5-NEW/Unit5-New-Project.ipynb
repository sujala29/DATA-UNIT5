## Clarity

Its one of the important principles of effective data communication is clarity.

The message you are trying to convey should be clear and easy to understand, without any ambiguity or room for misinterpretation.

This can be achieved by using simple language, avoiding technical confusion, and presenting the data in a visually appealing manner.

## Relevance: 

When communicating data, it's important to ensure that it is relevant to the audience you are addressing. 

This means understanding their needs, interests, and level of knowledge, and tailoring your message accordingly.

## Context:

It's important to provide context and background information that helps the audience understand the meaning and significance of the data, as well as how it relates to the overall message.

This includes information about the sources of the data, any limitations or biases, and any trends or patterns that are relevant.

# Basics of Visual Perception

## Visual Perception:

Visual perception refers to the process by which the brain interprets and makes sense of information received from the eyes.

### Order:

Order refers to the arrangement of objects and elements within a visual scene, and how they are organized and structured to convey meaning. 

In visual perception, order helps the brain to understand the relationships between objects and to recognize patterns and structure in the environment.

### Hierarchy:

Hierarchy refers to the way in which elements in a visual scene are arranged in terms of their importance or prominence.

In visual perception, hierarchy helps the brain to understand which elements are the most significant and to focus attention on the most relevant information.

### Relationships:

Relationships refer to the connections and associations between different elements in a visual scene.

In visual perception, relationships help the brain to understand how objects and elements are related to one another and to build a coherent representation of the visual environment.

### Convention:

Convention refers to the common practices and agreed-upon rules for arranging and presenting visual information.

In visual perception, convention helps the brain to quickly and easily recognize familiar patterns and structures, and to understand how information is being presented.

Conventions also help to guide visual perception and make it more efficient and predictable.

## Misleading charts:

### A political candidate might use a bar chart to show a downward trend in crime rates, but the chart might be misleading because it starts at a high value and ends at a low value, making the change appear more significant than it actually is.

# 2.Question : Are there any gender-based salary disparities in different industries and education levels?

# Plots: Heatmap/Cluster bar chart

```python


# pivot the dataframe to create a gender by industry table of mean salaries
pivot_df = df_salary.pivot_table(index='gender', columns='industry', values='salary', aggfunc='mean')

# create heatmap with seaborn
sns.heatmap(pivot_df, cmap='YlGnBu', annot=True, fmt='.0f', cbar_kws={'label': 'Mean Salary'})

# set plot labels and title
plt.xlabel('Industry')
plt.ylabel('Gender')
plt.title('Mean Salary by Gender and Industry')

# display the plot
plt.show()
```

# From the plot, we can observe that there are certain industries where there is a significant gender-based salary disparity, such as the finance and insurance industry, where males tend to earn a higher salary than females.
# On the other hand, in some industries, such as the information industry, there is no significant difference in salaries between genders. Additionally, we can see that generally, individuals with higher education levels tend to earn a higher salary in all industries and for both genders.

# QUESTION 3: How does industry experience impact salary across different genders and education levels?

# Plots:  scatter plot/line plot

```python


# group the data by gender and education level
grouped = df_salary.groupby(['gender', 'education'])

# plot the salary as a function of industry experience for each group
fig, ax = plt.subplots(figsize=(10, 6))
for name, group in grouped:
    ax.plot(group['industry_experience'], group['salary'], label=name)
    
# add labels and legend
ax.set_xlabel('Industry Experience')
ax.set_ylabel('Salary')
ax.set_title('Salary vs Industry Experience by Gender and Education Level')
ax.legend()
plt.show()
```

# Analysis: From the plot, we can observe that the salary increases as industry experience increases for all three levels of education. We can also notice that the slope of the line for masters degree is steeper than high scool and some degree ," indicating that having a higher degree leads to a higher salary for the same amount of industry experience. Additionally, we can see that the salary for females is consistently lower than males for each level of education and industry experience.
